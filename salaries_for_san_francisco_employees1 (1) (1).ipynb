```python
## OUTLIER DETECTION AND HANDLING

from scipy.spatial.distance import mahalanobis
import numpy as np
# Selecting relevant features
features = data[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']]
mean = features.mean(axis=0)
cov_matrix = np.cov(features.values.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)

# Calculate Mahalanobis distance
data['Mahalanobis'] = features.apply(lambda x: mahalanobis(x, mean, inv_cov_matrix), axis=1)

# Threshold for outliers
threshold = data['Mahalanobis'].quantile(0.99)
outliers = data[data['Mahalanobis'] > threshold]
data_cleaned=data[data['Mahalanobis']<=threshold]

print(f'Original dataset size: {data.shape[0]}')
print(f'Number of outliers: {len(outliers)}')
print(f'Cleaned dataset size: {data_cleaned.shape[0]}')

```

```python
from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
data['Outlier'] = iso_forest.fit_predict(features)

# Remove outliers where Isolation Forest flagged them as -1
Outlier = data[data['Outlier'] == -1]
cleaned_data_set=data[data['Outlier']!=-1]

print(f'Original dataset size: {data.shape[0]}')
print(f'Total Outliers Detected: {len(Outlier)}')
print(f'Cleaned dataset size: {cleaned_data_set.shape[0]}')
```

```python
from scipy.stats import zscore

# Compute Z-score for TotalPayBenefits
data['Zscore_TotalPayBenefits'] = zscore(data['TotalPayBenefits'])
#outliers based on z_score
z_outliers = data[(data['Zscore_TotalPayBenefits'] >3) | (data['Zscore_TotalPayBenefits'] <-3)]

# Remove outliers (Z-score > 3 or < -3)
data_cleared = data[(data['Zscore_TotalPayBenefits'] <= 3) & (data['Zscore_TotalPayBenefits'] >= -3)]

print(f'Original dataset size: {data.shape[0]}')
print(f'Outliers based on Z-Score: {len(z_outliers)}')
print(f'Cleaned dataset size: {data_cleaned.shape[0]}')
```

```python
Q1 = data['TotalPayBenefits'].quantile(0.25)
Q3 = data['TotalPayBenefits'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
#outliers count
outliers_count=data[(data['TotalPayBenefits'] <lower_bound) | (data['TotalPayBenefits'] > upper_bound)]

# Remove outliers
data_cleaned = data[(data['TotalPayBenefits'] >= lower_bound) & (data['TotalPayBenefits'] <= upper_bound)]

print(f'Original dataset size: {data.shape[0]}')
print(f'Outliers : {len(outliers_count)}')
print(f'Cleaned dataset size: {data_cleaned.shape[0]}')
print(outliers_count)
```

```python
columns_to_check = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPayBenefits']

for col in columns_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data= data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]

print(f'Cleaned dataset size after removing outliers from multiple columns: {data.shape[0]}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor


results = []
# Dictionary to store models and results
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}

# Train and Evaluate Models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    results.append({'Model': name,'MAE':mae,'MSE':mse,'RMSE': rmse, 'R2 Score': r2})

    print(f"\n{name} Performance:")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"RÂ² Score: {r2:.2f}")
result_data=pd.DataFrame(results)
result_data=result_data.sort_values("R2 Score",ascending=False)
result_data

```
